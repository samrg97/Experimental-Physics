## Introduction

In this experiment, photons are sent through a cell containing two isotopes of Rubidium: $\text{Rb}^{85}$ and $\text{Rb}^{87}$. As the temperature of the Rubidium increases, the density of the Rubidium increases which means there is an increased likelihood that a photon will get absorbed by valence electrons in the Rubidium atoms. It should follow that as the density of Rubidium increases, the photon flux that makes it through the cell of Rubidium decreases. This relationship follws an exponential decay.

$$I = I_0e^{-\sigma\ell\rho}$$

Where $I$ is the photon flux that makes it through the Rubidium, $I_0$ is the incident photon flux, $\sigma$ is the cross section of a Rubidium atom, $\ell$ is the path length of the photons, and $\rho$ is the density of Rubidium.

The total angular momentum of a Rubidium electron comes from spin of the electron, nuclear spin, and angular momentum:

$$F = I + J + S$$

Where $F$ is total angular momentum, $I$ is nuclear spin, $J$ is angular momentum, and $S$ is spin of the electron. When an external magnetic field is applied, the $F$ levels are split according to their quantum projection number $M$ due to the Zeeman effect. When the external magnetic field is weak, the energy change due to a Zeeman interaction is proportional to the magnetic field.

$$\Delta E = g_f\mu_0 B \Delta M$$

Where $\Delta E$ is the change in energy, $g_f$ is Lande's $g$ factor, $\mu_0$ is the Bohr magneton, $B$ is the strength of the external magnetic field, and $\Delta M$ is the change in projection quantum number, which is -2,-1,0,1,2 in the $F=2$ state and -1,0,1 in the $F=1$ state.

If the polarization of the photons is controlled, then the photons incident to the Rubidium gas "pump" valence electrons to higher energy levels because the fixed polarization fixes the change in projection quantum number $M$ to be +1. During the emission process, the polarization of those photons is not controlled, which means the electrons have equal chances at changing their projection quantum numbers by 0 or +/- 1. Because these changes have equal probability of occuring, the average change in $M$ is 0, so the electrons tend to stay in their higher energy levels. When the external magnetic field responsible for the Zeeman splitting is turned off, the energy levels mix back into what they were before the field was introduced and the electrons are forced to return to their original energy levels in equilibrium.

When the resonance condition is met while an RF frequency creates the sweeping magnetic field, the following equation holds:

$$hf_{rf} = \mu_Bg_fB$$

Where $h$ is Planck's constant, $f_{rf}$ is the resonant frequency, $\mu_B$ is the Bohr Magneton, $g_f$ is Lande's factor for a particular isotope, and $B$ is the strength of the sweeping magnetic field. Energy can be written as Planck's constant times frequency and $\Delta M$ is either 0 or +/- 1 so this equation is really just the previous equation but specifically for the resonance condition.

Transient Effects:

When the RF field is turned on and off, transient effects can be observed. Right after the RF field is turned back on, the electrons - that were previously upgraded to higher energy levels because they absorbed photons - are trying to return to their equilibrium energy levels but are constantly overshooting their equilibrium energy levels less and less over time until they all eventually reach equilibrium. This overshooting motion is oscillation and the period of this oscillation is inversely proportional to the amplitude of the RF field because the Lamor frequency (the frequency of precession of total angular momentum $F$) is proportional to the strength of the RF field.

$$f = g_f\frac{\mu_0}{\hbar}B_{rf}$$

In this equation, $f$ is the frequency of the oscillation from the electrons attempting to re-establish equilibrium, $g_f$ is the Lande's factor of a specific isotope, $\mu_0$ is the Bohr magneton, $\hbar$ is Plank's constant $h$ divided by $2\pi$, and $B_{rf}$ is the strength of the RF field.

## Analysis

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.odr as odr

from IPython.display import Image

import matplotlib.patches as mpatches
```

```python
def linear(param,x):
    return param[0]*x + param[1]

def C_to_K(temps_C):
    temps_k = []
    for i in range(0,len(temps_C)):
        temps_k.append(temps_C[i]+273.15)
    return temps_k

def temp_to_dens(temps,table):
    dens = []
    for i in range(0,len(temps)):
        for t,d in table.items():
            if abs(temps[i]-t) <= 5:
                dens.append(d)
    return dens
```

```python
# -------------------------------------------------- Part I -----------------------------------------------------------------
table1 = {290: 3.3*pow(10,15), 300: 1.1*pow(10,16), 310: 2.9*pow(10,16), 320: 7.5*pow(10,16), 330: 1.8*pow(10,17), 340: 4.3*pow(10,17), 350: 8.3*pow(10,17), 360: 1.5*pow(10,18), 370: 3.7*pow(10,18), 380: 6.3*pow(10,18), 390: 1.2*pow(10,19), 400: 2.4*pow(10,19)}
length = 0.025

temps_C = [37,47,57,67,77,87] # in Celsius
temps_K = C_to_K(temps_C)

densities = np.array(temp_to_dens(temps_K,table1))
voltages = [0.696,0.628,0.519,0.375,0.259,0.200,0.175]
for i in range(0,len(voltages)-1):
    voltages[i] = voltages[i] - voltages[-1]
voltages = voltages[:-1]
dv = 0.001
volt_log = np.log(voltages)
d_logv = []
for i in range(0,len(voltages)):
    d_logv.append(dv/voltages[i])

# Plot the data and make title and labels
plt.scatter(densities,volt_log)
plt.errorbar(densities,volt_log,yerr=d_logv,fmt="none",color='red',elinewidth=2,linewidth=0,marker='.',markersize=10,label='Error')
plt.xlabel("Atomic Density [atoms/m^3]")
plt.ylabel("log(V)")
plt.title("Figure 1.1: Rubidium Absorption as Temperature Varies")
plt.grid()

# Find a Linear Fit for the data
linear_fit = odr.Model(linear)          
noised_data = odr.RealData(densities,volt_log)  
regressed_model = odr.ODR(noised_data,linear_fit,beta0=[-2.1e-18,-0.67])  
regressed_output = regressed_model.run() 
regressed_output.pprint()   

X = np.linspace(0,1.75*pow(10,18),6)

fitted_params = regressed_output.beta    
param_sd = regressed_output.sd_beta    
chi_squared = str(round(regressed_output.res_var,3))  
fitted_curve = linear(fitted_params,X) 

plt.plot(X,fitted_curve,'orange',label='Linear Fit')
plt.text(1,-3,s="Chi-Squared: " + chi_squared)

plt.show()

# Calculate cross section from slope of linear fit
cross_section = -fitted_params[0]/length
d_cs = param_sd[0]/length
print("Cross section: " + str(round(cross_section,18)) + " m^2 \u00B1 " + str(round(d_cs,19)) + " m^2")
```

The voltage output by the photodetector was recorded at several temperatures, and each temperature corresponds to the density of the Rubidium gas. The voltages are proportional to the photon flux that reaches the photodetector, and since this flux decreases exponentially as the density of the gas increases, the natural log of the voltage is inversely proportional to the density. When the natural log of the voltage is plotted as a function of atomic density, the magnitude of the slope of the linear fit is the cross section of a Rubidium atom multiplied by the path length of the photons (the path length is known: 2.5 cm).

The uncertainty in the cross section was calculated through error propagation starting from defining the uncertainty in the voltage output from the photodetector. The uncertainty in voltage was plugged into ODR and ODR output the uncertainty in the slope. The slope is the cross section multiplied by the negative path length, so the uncertainty in the cross section is the uncertainty in slope divided by the negative path length.

The chi-squared value of the data in Figure 1.1 is really low and the experimentally determined cross section is an order of magnitude smaller than the literature value. A potential source of error for this part of the experiment is the temperature through the whole cell containing the Rubidium might not have had a uniform temperature, so the density of the gas is different enough from the densities given to produce this different in cross section. There also might have been accounted for background from a few photons coming from the lights in the lab travelling under the black cover on the lab apparatus and reaching the photodetector.

The data in Figure 3.1 is broken up into 4 groups: two different isotopes at two different frequency ranges. The horizontal sweep range had to be changed for higher frequencies in order to see the peaks of the isotopes, so the sweep range was not consistent for all data points. The four data points at the low frequencies have the same sweep range and the three data points at the high frequencies have the same sweep range, so linear fits were found for high and low frequencies separately to keep the sweep ranges consistent for each linear fit. To find the g-factor ratio, the g-factor ratio for low frequencies was found separately from the g-factor ratio for high frequencies. A weighted average of the two ratios was taken to find the g-factor ratio of the whole set of data.

The uncertainty in the g-factor ratio was calculated with error propagation starting from knowing the uncertainty of the times at which the isotopes' peaks occurred on the oscilloscope and the uncertainty of the RF frequency. The oscilloscope kept track of data points in 2 ms intervals, so the uncertainty in time is 2 ms. The function generator that produced the current that created the RF field output its frequency with a certainty out to the nearest Hz, so the uncertainty in the frequency is 1 Hz. Those uncertainties were plugged into ODR when finding the linear fits for the data, and ODR output the uncertainties in the slopes of each linear fit. The uncertainties in the ratios for low and high frequencies were found using the formula for error propagation when dividing uncertain values. The uncertainty of the final ratio was found by taking a weighted average of the two uncertainties, similar to how the final ratio was found by taking a weighted average of the two ratios.

The square wave is the RF field turning on and off, and the other wave drops down when the RF field is turned off and increases logarithmically when the RF field is turned back on. Optical pumping only occurs when the RF field is off and when this happens, the transmitted light intensity increases exponentially. The signal then saturates after a period of time while the RF field is off because almost all the light going into the gas is coming out and reaching the photodetector; the light intensity that reaches the photodetector cannot be greater than the light intensity goin into the gas. This rate of increase in light intensity corresponds to the time constant that describes the exponential growth, and the time constant is proportional to the intensity of the incoming light intensity. If the incoming light intensity increases, then it will take a longer amount of time from the RF field being turned off for the signal to reach maximum light intensity. After the RF field is turned on, the signal oscillates because the electrons in the gas are returning to equilibrium since the Zeeman effect that split the energy levels is no longer present. As the electrons attempt to reach equilibrium, they overshoot the drop in energy needed to reach equilibrium and need to attempt to move back up to the equilibrium level they skipped over, but they will overshoot again slightly and need to move back down again. The electrons overshoot reaching their equilibrium levels less and less over time until they eventually all reach equilibrium. This motion of the electrons overshooting and undershooting their equilibrium levels is an oscillation, which is what the oscilloscope signal is showing. When the signal decays back to a flat line, that is when all the electrons have reached their equilibrium levels. The frequency of this oscillation is proportional to the amplitude of the RF field. The frequency depends on the amplitude of the RF field because the precession frequency of the total angular momentum $F$ depends on the strength of the RF field.
