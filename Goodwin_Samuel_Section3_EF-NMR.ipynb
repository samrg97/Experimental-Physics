## Introduction

Protons have a spin of +1/2, which means they have a magnetic moment and will, therefore experience a torque when exposed to a magnetic field. Because Earth produces a magnetic field, water molecules experience atorque in order for the molecule to line up with Earth's magnetic field. If an external magnetic field is applied to water molecules perpendicular to Earth's magnetic field, a resultant magnetic field will form at some angle from Earth's magnetic field, producing a magnetization. If the external field is removed quickly, the resultant magnetic field will shrink down to Earth's magnetic field but the magnetization lasts temporarily and precesses around Earth's magnetic field with frequency $\omega$.

$$\omega = \lambda B_e$$

where $\lambda$ is a scalar and $B_e$ is the strength of Earth's magnetic field.

Curie's Law can be used to find the equilibrium magnetization of a sample of magnetic moment.

$$M_0 = \frac{N(\frac{l+1}{l})\mu^2}{3kT}B$$

where $N$ is the number of magnetic moments per volume, $l$ is the spin quantum number for protons (1/2), $\mu$ is the magnetic moment for the spin of the proton, $B$ is the applied magnetic field, $k$ is the Boltzman Constant, and $T$ is the temperature in Kelvin.

The magnetization of the water approaches equilibrium exponentially with time constant $T_1$ (the spin-lattice relaxation time).

$$M(t) = M_0(1 - e^{t/T_1})$$

I can rearrange this equation to look like this:

$$M_0 - M(t) = M_0e^{-t/T_1}$$

Then, I can take the natural log of each side:

$$\ln(M_0-M(t)) = \ln(M_0) - \frac{t}{T_1}$$

By doing this, I have manipulated the exponential decay of magnetization equation to look like a linear function with a slope based solely on relaxation time $T_1$.

$$\text{slope}: -\frac{1}{T_1}$$
$$\text{y-intercept}: \ln(M_0)$$

## Analysis

- A plot relating amplitude of free precession and polarization time for both signals. 
- A curve fit to the plot relating amplitude of free precession and polarization time for both signals. 
- A determination of the relaxation time of room temperature water. 
- A plot relating amplitude of free precession versus polarization current. 
- Discussion of the results and their relationship to the Curie Law. 

```python
import matplotlib.pyplot as plt
import math
import numpy as np
import scipy.odr as odr
```

```python
def convert_to_log(values):
    for i in range(0,len(values)):
        values[i] = math.log(values[i])

def linear_func(params,x):
    return params[0]*x + params[1]

error_x_t = 0.05
error_x_c = 0.01
error_y = 0.00005
```

```python
times = np.array([13.0,5.0,4.0,3.0,2.0,1.0,0.5])
w_t_peaks = [0.012,0.0108,0.0108,0.010,0.0092,0.0076,0.0064]
for i in range(1,len(w_t_peaks)):
    w_t_peaks[i] = w_t_peaks[0] - w_t_peaks[i]
w_t_peaks.pop(0)
times_copy = np.delete(times,0)
convert_to_log(w_t_peaks)

plt.scatter(times_copy,w_t_peaks)
plt.errorbar(times_copy,w_t_peaks,yerr=error_y,xerr=error_x_t,fmt="none",color='red',elinewidth=2,linewidth=0,marker = '.',
             markersize = 10,label = 'Background Noise')
plt.title('Figure 1.1: Relating Free Precession to Polarization Time')
plt.xlabel('Polarization Time [s]')
plt.ylabel('ln(M_0 - M(t))')

linear_t = odr.Model(linear_func)
noised_data_t = odr.RealData(times_copy, w_t_peaks, sx = error_x_t, sy = error_y)
regressed_model_t = odr.ODR(noised_data_t,linear_t,beta0=[1,-5])
regressed_output_t = regressed_model_t.run()
regressed_output_t.pprint() 

fitted_params_t = regressed_output_t.beta
param_sd_t = regressed_output_t.sd_beta
chi_squared_t = str(round(regressed_output_t.res_var,1))
fitted_curve_t = linear_func(fitted_params_t,times_copy)

plt.plot(times_copy,fitted_curve_t)
plt.text(3,-6,s='Chi Squared: '+ chi_squared_t)

plt.show()

print("Relaxation Time of Water:  T_1 =", round(-1/fitted_params_t[0],2), 's \u00B1', round(param_sd_t[0],2), 's')
```

```python
currents = np.array([3.0,2.5,2.0,1.5,0.5])
w_c_peaks = [0.0116,0.0099,0.0092,0.0075,0.0049]

plt.scatter(currents,w_c_peaks)
plt.errorbar(currents,w_c_peaks,yerr=error_y,xerr=error_x_c,fmt="none",color='red',elinewidth=2,linewidth=0,marker = '.',
             markersize = 10,label = 'Background Noise')
plt.title('Figure 1.2: Relating Free Precession to Polarization Current')
plt.xlabel('Polarization Time [s]')
plt.ylabel('Ampltiude of Free Precession')

linear_c = odr.Model(linear_func)
noised_data_c = odr.RealData(currents, w_c_peaks, sx = error_x_c, sy = error_y)
regressed_model_c = odr.ODR(noised_data_c,linear_c,beta0=[1,-6])
regressed_output_c = regressed_model_c.run()
regressed_output_c.pprint() 

fitted_params_c = regressed_output_c.beta
param_sd_c = regressed_output_c.sd_beta
chi_squared_c = str(round(regressed_output_c.res_var,1))
fitted_curve_c = linear_func(fitted_params_c,currents)

plt.plot(currents,fitted_curve_c)
plt.ylim(0,0.014)

plt.show()
print('Chi Squared: ', chi_squared_c) # plt.text() does not want to play nice
```

Discussion:

My data shows that amplitude of free precession increases linearly with current. Given that amplitude of free precession is proportional to magnetization and current is proportional to the strength of the magnetic field, my data satisfies Curie's Law, which states that equilibrium magnetization increases linearly with magnetic field strength. This also means the slope of my plot is proportional to $\frac{N\mu^2}{kT}$ where $N$ is the number of magnetic moments per volume, $\mu$ is the magnetic moment due to spin of a proton, $k$ is the Boltzmann constant, and $T$ is room temperature in Kelvin.
