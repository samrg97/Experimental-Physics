## Introduction

Circuits can model differential equations by measuring rates of change in voltage. The main parts of the circuit components that can measure these rates of change are operational amplifiers (or "op-amps" for short). The important pins on the op-amp for performing this lab are the two voltage inputs - inverting (-) and non-inverting (+), two power inputs (one positive and one negative), and an output voltage. The voltage inputs do not take in current but know the voltage difference between the two inputs.

![OpAmp](op-amp.png)

One circuit that is useful for this experiment is a summation amplifier (which will be referred to as an "adder"), which takes in parallel voltage input through the inverting input and outputs the sum of the voltages.

![AdderDiagram](adder.png)

$$-\frac{R_{f0}}{R_1}V_1 - \frac{R_f}{R_2}V_2 - \cdots - \frac{R_f}{R_n}V_n = V_0$$

This equation was derived from Kirchoff's and Ohm's Laws.

The inverter circuit (or "inverter") takes a voltage input and inverts it.

![InverterDiagram](inverter.png)

The relationship between the input and output voltages is shown by the equation below.

$$-\frac{R_{f0}}{R_{in}}V_{in} = V_{out}$$

The differentiating circuit (or "differentiator") takes an input voltage and outputs a voltage that satisfies the equation below:

![DiffDiagram](differentiator.png)

$$V_{out} = V_0(0) - RC\frac{dV_{in}(t)}{dt}$$

The integrating circuit (or "integrator") takes an input voltage and outputs a voltage that satisfies the equation below:

![IntDiagram](integrator.png)

$$V_{out} = V_0(0) - \int_0^{t'} \frac{V_{in}(t')}{RC}dt'$$

Part II:

To find the spring constant $k$, I used the integrator formula twice to get an equation that looks like $x$ as a function of $x''$. The constant in terms of $R_1$, $R_2$, $C_1$, $C_2$ in front of the $x''$ term is $k$. However, the feedback resistor $R_f$ and the input resistors $R_{in}$ are not equal to each other, so the constant in front of $x''$ has to be scaled by $\frac{R_f}{R_{in}}$ to get the $k$ spring constant of the "oscillator".

$$k = (R_f/R_{in})(\frac{1}{R_1C_1})(\frac{1}{R_2C_2})$$

![CircuitDiagram2](circuit_diagram2.jpg)

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.odr as odr
import pandas as pd
import matplotlib.patches as mpatches
```

```python
def osc_decay(param,t):
    # 0: amplitude
    # 1: decay constant
    # 2: frequency
    # 3: phase angle
    # 4: voltage offset
    return param[0]*np.exp(-param[1]*t/2)*np.cos(param[2]*t + param[3]) + param[4]
```

```python
# --------------------------------------------- Part II -----------------------------------------------------------
time2 = np.linspace(0,1693,2200)

data2 = pd.read_csv("ALL0009/F0009CH2.csv")
voltage2 = data2.iloc[0:2200,4:5].values.reshape(2200,)

plt.plot(time2,voltage2)
plt.ylabel("Voltage [V]")
plt.xlabel("Time [ns]")
plt.title("Figure 2.1: Simple Harmonic Oscillator")

R1 = 5047
R2 = 5026
Rf = 1.2
Rin = 1.15
dR = 1
C1 = 11e-9
C2 = 11e-9
dC = 1e-9

int1 = 1/(R1*C1)
int2 = 1/(R2*C2)
adder = Rf/Rin
numer = np.sqrt(int1*int2*adder)
denom = 2*np.pi
f = (numer/denom)*pow(10,-9)

dint1 = (1/(R1*C1)**2)*(R1*C1*(dR/R1 + dC/C1))
dint2 = (1/(R2*C2)**2)*(R2*C2*(dR/R2 + dC/C2))
dadder = adder*((dR/Rf)+(dR/Rin))
df = (0.5*(dint1*dint2*dadder)**-0.5)*pow(10,-9)

# Find a Linear Fit for the data
sin_decay = odr.Model(osc_decay)          
my_data = odr.RealData(time2,voltage2,sx=np.std(time2),sy=np.std(voltage2))  
regressed_model = odr.ODR(my_data,sin_decay,beta0=[4.57603062e-01,  3.24864146e-04,  4.30363723e-02,  4.22288990e+00,
 -9.33375419e-01])  
regressed_output = regressed_model.run() 
regressed_output.pprint()   

#X = np.linspace(-50,1760,5000)

fitted_params2 = regressed_output.beta    
param_sd2 = regressed_output.sd_beta    
chi_squared = str(round(regressed_output.res_var,3))  
fitted_curve = osc_decay(fitted_params2,time2) 

plt.plot(time2,fitted_curve,'orange',label='Linear Fit')
blue_patch = mpatches.Patch(color='blue', label='x(t)')
orange_patch = mpatches.Patch(color='orange', label='Sinusoidal Decay Fit')
plt.legend(handles=[blue_patch,orange_patch])

plt.show()
print("Chi-Squared: " + chi_squared)
print("Frequency: f = " + str(round(fitted_params2[2]/(2*np.pi),2)) + " ns^{-1} \u00B1 " + str(round(param_sd2[2]/(2*np.pi),5)) + " ns^{-1}")
print("Theoretical frequency: f = " + str(round(f,7)) + " ns^{-1} \u00B1 " + str(round(df,14)) + " ns^{-1}")
print("Damping constant: " + str(round(fitted_params2[1],5)) + " ns^{-1} \u00B1 " + str(round(param_sd2[1],6)) + " ns^{-1}")
```

I found the theoretical frequency $f$ taking the square root of spring constant $k$ and dividing by $2\pi$ to convert angular frequency to frequency. The data from the oscilloscope was exported to a csv file and then pandas was used to parse the data into numpy arrays. ODR was used to find a sinusoidal decay fit for the data. Five parameters were used to fit the data: amplitude, decay constant, angular frequency, phase shift, and voltage offset.

An uncertainty in the resistors and capacitors was estimated. From there, error is propogated to find the uncertainty in frequency using the error propogation formula for multiplication and calculus. Uncertainty in the frequency fit and damping constant fit is calculated using ODR.

The reduced chi-squared of the fit to the data is very small (less than 1), which means the fit is over-constrained. Sources of error include the function generator behaving strangely and the ciruit did not use recommended capacitors. The function generator was supposed to produce square waves, but the pulse read on the oscilloscope was not perfectly square, and it was recommended to use capacitors in the microFarad range but the circuit used to collect this data uses 11 nF capacitors. The ciruit might not have worked as intended because of using these capacitors. Also, the fit to the data is slightly incorrect, which is known because the iteration limit for finding a fit was reached. A substitute for ODR may have produced a better fit.
