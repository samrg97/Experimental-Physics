## Introduction

Gamma-rays are high-energy photons that are typically generated through the gamma-decay of a heavy nucleus. When the energy spectra of emitted gamma-rays is measured, the rightmost-peak on the spectrum is the energy of the emitted gamma-rays. This peak is called the photopeak of the source of radiation. Gamma-ray emission follows the inverse square law, which means the number of detected gamma-rays decreases as a function of one over the distance of detection squared.

$$counts \propto \frac{1}{r^2}$$

Gamma-rays can sometimes pass through materials. The probability that a gamma-ray can pass through a material is called the transmission probability, which can be solved for by looking at the number of incident photons and the number of photons that make it through the material the photons are aimed at.

$$T = \frac{I}{I_0}$$

In this equation, $T$ is the transmission probability, $I$ is the number of photons that pass through the material, and $I_0$ is the number of photons that are aimed at the material.

Transmission probability can also be written as a function of the depth of the material and the attenuation coefficient of the material for a given incident photon energy.

$$T = e^{-\mu x}$$

Where $T$ is the transmission probability, $\mu$ is the attenuation coefficient of the material (which also depends on the energy of the incident photons), and $x$ is the distance the photons have to travel through the material in order to completely pass through it.  Photons are less likely to pass through materials with higher attenuation coefficients and the more material a photon is required to pass through, the less likely that photon is to completely pass through it.

Photons interact with matter in different ways depending on the energy of the photons. In this experiment, the main mechanism of photon-electron interaction that is observed is Compton scattering, which is when medium-level photons (about less than 2 MeV) collide with electrons inelastically. Conservation of energy and momentum exists in this interaction, and the energy of the gamma rays before and after the collision can be measured to find the change in momentum - and therefore the mass - of the electrons involed in the collision.

$$\frac{1}{E'} = \frac{1}{mc^2}(1-\cos\theta) + \frac{1}{E}$$

In this equation describing the conservation of energy and momentum of the photon and electron, $E'$ is the energy of the photon after it collides with the electron, $mc^2$ is the rest energy of the electron, $\theta$ is the angle at which the photon scatters after hitting the electron, and $E$ is the energy of the photon before it collides with the electron.

The differential cross-section for Compton Scattering can be written as $\frac{d\sigma}{d\Omega}$ where $d\sigma$ is the cross-sectional area of the plane of initial electrons and $d\Omega$ is the solid angle the photons are scattered at after colliding with electrons. The differential cross-section can be written as a function of scattering angle and radius of the electron.

$$\frac{d\sigma}{d\Omega} = \frac{r^2}{2}[\frac{1+\cos^2(\theta)}{[1+\alpha(1-\cos(\theta)]^2}][1+\frac{\alpha^2(1-\cos(\theta))^2}{(1+\cos^2(\theta))[1+\alpha(1-\cos(\theta)]}]$$

Where $r$ is the classical radius of the electron, $\theta$ is the angle the photons are scatter at, and $\alpha$ is described with the following equation:

$$\alpha = \frac{E}{mc^2}$$

Where $E$ is the energy of the photons before they are scattered and $mc^2$ is the rest energy of an electron.

Another equation that describes the differential cross-section is shown below:

$$\frac{d\sigma}{d\Omega} = \frac{\Sigma^{'}}{n_{\theta}I\Delta\Omega\epsilon t}$$

Where $\Sigma^{'}$ is the integral of the photopeak, $I$ is the decays per second, $\epsilon$ is the efficiency of the detector, $\Delta\Omega$ is the solid angle of the detector, $t$ is the time over which each scan was taken, and $n_{\theta}$ is a function of other parameters:

$$n_{\theta} = \rho VN_A\frac{Z}{M}$$

Where \rho is the density of the scattering rod, $V$ is the volume of the scattering rod, $N_A$ is Avogadro's number, $Z$ is the atomic number of the element that makes up the scattering rod, and $M$ is the molar mass of the scattering rod.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.odr as odr
from scipy.integrate import simps
import matplotlib.patches as mpatches
```

```python
def inv_square(param,x):
    return (param[2]/(x**param[0])) + param[1]

def linear(param,x):
    return param[0]*x + param[1]

def gaussian(param,x):
    return param[0]*np.exp(-(x-param[1])**2/(2*param[2]**2))



def find_avg(counts1,counts2,counts3):
    counts_avg = []
    for i in range(0,len(counts1)):
        sum_ = counts1[i]+counts2[i]+counts3[i]
        counts_avg.append(sum_/3)
    return counts_avg

def find_std(counts1,counts2,counts3,counts_avg):
    counts_std = []
    for i in range(0,len(counts_avg)):
        term1 = (counts1[i]-counts_avg[i])**2
        term2 = (counts2[i]-counts_avg[i])**2
        term3 = (counts3[i]-counts_avg[i])**2
        variance = (term1+term2+term3)/2
        std = np.sqrt(variance)
        counts_std.append(std)
    return counts_std

def scatter_spectrum(file_data):
    counts = []
    for _ in range(12):
        next(file_data)
    for line in file_data:
        if line[0] != ' ':
            break
        else:
            # Here are the counts
            count = ""
            for c in line:
                if c == " " or c == "\n":
                    continue
                else:
                    count += c
            count = int(count)
            counts.append(count)
    return counts

def find_energy(X,counts,guess):
    gauss_fit = odr.Model(gaussian)          
    my_data = odr.RealData(X,counts)  
    regressed_model = odr.ODR(my_data,gauss_fit,beta0=guess)  
    regressed_output = regressed_model.run() 
    regressed_output.pprint()   

    fitted_params = regressed_output.beta    
    param_sd = regressed_output.sd_beta    
    fitted_curve = gaussian(fitted_params,X)

    result = simps(fitted_curve,X)
    plt.plot(X,fitted_curve,'orange',label='Linear Fit')
    return (fitted_params[1],1/(fitted_params[0]*np.sqrt(2*np.pi)),result)
```

```python
#------------------------------------ Inverse Square Relationship --------------------------------------------------
background = 9104

dist = np.array([0.031,0.005,0.116,0.074,0.091,0.405,0.293])
dist_unc = np.array([0.003,0.001,0.002,0.001,0.002,0.002,0.002])
dist_std = 0.15

counts1 = np.array([235095,637967,46038,89695,63767,8042,11976])
counts2 = np.array([233688,638833,45669,89146,63385,8130,11940])
counts3 = np.array([234639,637524,46287,89071,63225,8053,12176])
counts1 = counts1-background
counts2 = counts2-background
counts3 = counts3-background

counts_avg = find_avg(counts1,counts2,counts3)
counts_std = find_std(counts1,counts2,counts3,counts_avg)


# Plot the data and make title and labels
plt.scatter(dist,counts_avg)
plt.errorbar(dist,counts_avg,xerr=dist_unc,yerr=counts_std,fmt="none",color='red',elinewidth=2,linewidth=0,marker='.',markersize=10,label='Error')
plt.xlabel("Distance from Source [m]")
plt.ylabel("Counts")
plt.title("Figure 1.1: Gamma Counts as Distance Changes")
plt.grid()

# Find a Linear Fit for the data
inv_square_fit = odr.Model(inv_square)          
my_data = odr.RealData(dist,counts_avg,sx=dist_std,sy=counts_std)  
regressed_model = odr.ODR(my_data,inv_square_fit,beta0=[1.5, 0.08578745, 0.20314911])  
regressed_output = regressed_model.run() 
regressed_output.pprint()   

X = np.linspace(0.005,0.43,50)

fitted_params = regressed_output.beta    
param_sd = regressed_output.sd_beta    
chi_squared = str(round(regressed_output.res_var,3))  
fitted_curve = inv_square(fitted_params,X) 

plt.plot(X,fitted_curve,'orange',label='Linear Fit')
plt.ylim([-0.1e7,0.2e7])

plt.show()

print("Chi-Squared: ", chi_squared)
print("Distance-Count Relationship: 1/r^(" + str(round(fitted_params[0],2)) + "\u00B1" + str(round(param_sd[0],0)) + ")")
```

The counts of gamma rays with no plates is the number of incident photons. When plates were added, the gamma counts measured on the other side of the plates are the number of transmitted photons, so the transmission probability can be written as number of transmitted photons divided by number of incident photons. Since transmission probability is an exponential decay as depth of material increases, the natural log of transmission probability is equal to the negative linear attenuation coefficient multiplied by the depth of the material. The natural logs of the transmission probabilities are plotted as a function of depth of the aluminum plates, and the slope of the linear fit is the negative linear attenuation coefficient. The uncertainty in the depth of the plates was picked based on how certain measurements with a ruler are and uncertainty in incident and transmitted photons is the square root of number of incident and transmitted photons respectively. When converting the transmission probabilities to the natural logs of the transmission probabilities, uncertainty was carried over using calculus. The derivative of the natural log of transmission probability is 1 divided by the transmission probability, so uncertainty in natural log of transmission probability is this quantity multiplied by uncertainty in transmitted photons divided by uncertainty in incident photons. From here, ODR took care of error propagation for me, and the uncertainty in slope was the uncertainty in the linear attenuation coefficient. The chi-squared value for this plot is much too small to be acceptable.  Error in this part of the experiment could have been caused by the plates sometimes not placed perfectly orthogonal to the path of the photons. Also, the heights of the source, plates, and detector were not as even as they could have been.

The energy spectra for each angle was parsed with code. The bins of the histogram (energy spectra) were calibrated using the linear fit for the channels found in the very beginning of the experiment when Maestro 7 was calibrated. Once the counts for each energy bin were found, the energy spectra for each angle was plotted so that the photopeak for each angle could be found by just looking at the graph. Using ODR, the part of the spectrum that was determined to be a photopeak was fit to a Gaussian and the mean value in the Gaussian is the energy of the gamma-rays at that particulat angle and the standard deviation is the uncertainty in the mean. Once the gamma energies were found for each angle from 20 degrees to 80 degrees, the reciprocal of the energies were plotted as a function of one minus the cosine of the angle, and this plot was fit to a linear function. The slope of this linear fit is the reciprocal of the rest energy of an electron. Uncertainty in this slope was calculated by plugging the standard deviations from the Gaussian fits into ODR and letting ODR find the uncertainty in slope. From here, the uncertainty in mass of the electron was found by using the experimentally determined mass of the electron, uncertainty in the slope of the linear fit, and calculus.  The chi-squared value for the plot used to find mass of the electron is far too small to be acceptable. Error in this part of the experiment most likely originates from the width of the gap the gamma rays travelled through being too large. In order to have minimum exposure to radiation, the gap made from two small lead bricks was made hastily when measurements were taken and the width of this gap should have been a little smaller than what it actually was. The gap needs to be small so that all photons leaving the source scatter off the aluminum rod.
