## Introduction

According to Biot-Savart's Law, the strength of the magnetic field due to N turns in the loop of wire is:

$$B(z) = \frac{N\mu_0Ir'^2}{2(r'^2+z^2)^{3/2}}$$

Where $\mu_0$ is vacuum permeability, $I$ is current in the loop, $r'$ is the radius of the loop, and $z$ is distance along the z-axis assuming $z=0$ is at the center of the loop. In this experiment, there are 195 turns in each of two coils, so I can simplify this to one coil with 390 turns in my calculations.

In the first part of this experiment, I balanced out gravitational torque from a weight attached to the cue ball and magnetic torque due to the magnet embedded in the center of the cue ball. The gravitational torque is the cross product of displacement from the center of the cue ball to the center of the weight I added $\overrightarrow{r}$ and weight of the weight I added $m\overrightarrow{g}$. The magnetic torque of the ball is the cross product of the magnetic moment $\overrightarrow{\mu}$ and the magnetic field $\overrightarrow{B}$

$$\tau_{grav} = \overrightarrow{r} \times m\overrightarrow{g}$$
$$\tau_{mag} = \overrightarrow{\mu} \times \overrightarrow{B}$$

The angles of separation for the two different torques are the same, so at equilibrium I can use the equation:

$$\mu B = rmg$$

I can solve for $B$ as a function of $r$:

$$B = \frac{mg}{\mu}r$$

In the third part of this experiment, I am looking at harmonic oscillation of the rod due to the magnetic field producing a torque on the ball. Torque is the change in angular momentum over time.

$$\tau = \frac{dL}{dt}$$

Where $\tau$ is the net torque on the ball and $L$ is the angular momentum of the ball. Angular momentum can be written as the product of moment of inertia $I'$ and angular velocity $\omega$.

$$L = I'\omega$$

If I make subsitutions in my original equation for torque, and notice $\omega = \frac{d\theta}{dt}$, then I can say:

$$\overrightarrow{\mu}\times\overrightarrow{B} = \frac{d}{dt}(I'\omega)$$

Which can be rewritten to:

$$\mu B\sin\theta = I'\frac{d^2\theta}{dt}$$

Using the small angle approximation $\sin\theta = \theta$, I can rewrite my equation to look like a differential equation with a well-known solution.

$$\mu B\theta = I'\frac{d^2\theta}{dt}$$

The solution for this type of differential equation is $\theta(t) = A\cos(\omega t)$, so now the equation is:

$$\mu BA\cos(\omega t) = -I'A\omega^2\cos(\omega t)$$

From here, I can solve for $\omega^2$:

$$\omega^2 = \frac{\mu}{I'}B$$

And since the period of oscillation $T = \frac{2\pi}{\omega}$, I can rewrite my equation to solve for $T^2$:

$$T^2 = \frac{4\pi^2I'}{\mu B}$$

The moment of inertia of a solid sphere is $I' = \frac{2}{5}mr^2$ where $m$ is the mass of the sphere and $r$ is the radius of the sphere. I can substitute this expression for moment of inertia into the previous equation:

$$T^2 = \frac{8\pi^2mr^2}{5\mu B}$$

In the fourth part of this experiment, I am observing how the strength of the magnetic field affects precession frequency of a spinning magnet with a magnetic field passing through it. Over a short period of time, the spin angular momentum $L_s$ due to the ball spinning changes over some small angle $\Delta\phi$. If I call $\theta$ the angle from the z-axis to the spin axis, then the change in $L_s$ can be written as:

$$\Delta L_s = L_s\sin(\theta)\sin(\Delta\phi) = \Delta\phi\sin(\theta)$$

I used the small-angle approximation to rewrite the equation. If I divide each side of the equation by $\Delta T$, I get:

$$\frac{\Delta L_s}{\Delta T} = \frac{\Delta\phi}{\Delta T}L_s\sin\theta$$

If I let $\Delta T\xrightarrow{}0$ and I use the relationship $\tau = \frac{dL_s}{dt} = \mu B\sin\theta$, then I get:

$$\mu B\sin\theta = \Omega_pL_s\sin\theta$$

Where $\Omega_p = \frac{\Delta\phi}{\Delta T}$ is the frequency of precession. I can solve for frequency of precession now:

$$\Omega_p = \frac{\mu}{L_s}B$$

I know $L_s = I'\omega$, so I can solve for $\Omega_p$ in terms of the magnetic field and the frequency of the ball's spin (both of which are experimentally determined)

$$\Omega_p = \frac{\mu}{I'\omega}B$$

I can subsitute $I'$ with $\frac{2}{5}mr^2$ since the cue ball is a solid sphere:

$$\Omega_p = \frac{5\mu}{2mr^2\omega}B$$

In the fifth part of the experiement, one of the two loops has its current reversed in order to create a gradient. The force from the gradient of the magnetic field can be balanced out by adding weight to the other side of a balance beam to create a gravitational force to balance out the gradient force. This new magnetic field is:

$$B(z) = \frac{\mu_0Ir^2}{2(r^2+z^2)^{3/2}} - \frac{\mu_0Ir^2}{2(r^2+(d-z)^2)^{3/2}}$$

Since the magnetic field only changes along the z-axis, so to find the gradient, I just have to take the deriavtive of the magnetic field with respect to $z$. If I evaluate the gradient at $z = \frac{d}{2}$, I get the gradient halfway between the two coils.

$$\frac{dB}{dz} = \frac{3\mu_0Idr^2}{2(r^2+\frac{d^2}{4})^{5/2}}$$

Where $\mu_0$ is vacuum permeability, $r$ is the radius of the coils, $I$ is the current through the loop, and $d$ is the distance between the two coils.

The magnetic force acting on the dipole is given by:

$$F(z) = \mu \frac{dB}{dz}$$

Where $\mu$ is the magnetic moment of the magnet in the cue ball. This force can be balanced out by adding weight $mg$ to a lever, so I can say:

$$mg = \mu \frac{dB}{dz}$$

## Procedure

Required Materials:
-  Magnetic Torque apparatus

Part I: Static Magnetic Torque
- On the Magnetic Torque Apparatus, field gradient should be off, strobe light should be off, air should be on, field should be “up”. 
- Measure the diameter of the cue ball using the calipers.
- Measure the masses of the ball, rod, and weight separately using the triple beam balance.
- Measure the height of the weight and the length of the handle on the cue ball using a ruler. This is to help find the distance from the center of ball to the center of weight.
- Place the rod inside the handle and then place the weight at some point on the rod. Keep track of the position the weight is placed on.
- Put the cue ball on the air bearing and turn on the switch that says "AIR".
- The rod should tip over because the gravitational torque is stronger than magnetic torque (since there is no magnetic field present).
- Adjust the current on the Magnetic Torque Apparatus until you can tip the rod back to close to vertical and the rod does not tip back. Record this current.
- Move the weight to a different position on the rod and find the new minimum current needed to stop the rod from tipping. Repeat this with several weight positions.

Part II: Angular Dependence of Static Experiment
- Use a ruler to measure the difference in the outer and inner radii of the coils.
- Use a ruler to measure the distance from the center of the cue ball to the top of the upper coils.
- Place the center of the weight about 15 cm from the center of the cue ball. Record this position.
- Insert the rod in the handle of the cue ball (if it was not there already) and place the ball on the air bearing. Flip on the switch that says "AIR".
- Place a ruler on top of the coils across the diameter of the coils.
- Bring the rod to a point chosen on the ruler and find the current that acheives static equilbrium at that fixed angle. Record this current.
- Repeat the previous step with several other angles.

Part III: Harmonic Oscillation
- Remove the weight from the rod.
- Place the ball with the rod in the handle back on the air bearing. 
- Set the current to 1 A.
- Give the rod a small angular displacement from the vertical and allow it to oscillate.
- Use the stopwatch to record the time it takes to complete 20 cycles. Divide this time by 20 to find the period. Record the period for this current.
- Repeat the last two steps with several different currents up to 4 A. Get a lot of periods for smaller currents since the independent variable is 1/B.

Part IV: Precession Motion of a Spinning Sphere
- Turn off the current and turn on the "AIR" switch.
- Take the rod out of the ball and place it in the air bearing.
- Turn on the strobe light on the Magnetic Torque Apparatus and adjust its frequency to be about 5.5 Hz. Try spinning the ball and notice the ball will spin at the same frequency as the light flashes if it appears like the white dot on the handle is motionless. This might take a few (or several) practice spins to get right.
- As soon as the ball starts spinning at the same frequency as the light, turn on the current, increase it to 1 A, and start the stop watch.
- Time how long it takes for the handle to precess around the vertical and back to its starting point (when the timer was started). Record this period and this current.
- Repeat the last two steps with 1.5 A, 2.0 A, 2.5 A, 3.0 A, 3.5 A, and 4.0 A.

Part V: Net Force in a Field Gradient
- On the Magnetic Torque Apparatus, turn the strobe light off and the field gradient on.
- The balance beam should have two strings tied on each end. On one end of a string should be a Delrin Cup and the other string should have a counter weight.
- Attach the balance beam on top of the upper coil such that the Delrin Cup end is over the air bearing and the counter weight is hanging outside the coils.
- Measure the masses of the big and small brass balls using the triple balance beam.
- Attach the cue ball to the Delrin Cup such that the ball is not touching the air bearing. If the ball is touching the air bearing, add brass balls to the counterweight or retie the string.
- Add brass balls to the counterweight until equilibrium is reached. Use the height marker to keep track of what height the counterweight is at when equilibrium is reached.
- Place a small brass ball on the counterweight. There should no longer be equilibrium.
- Turn on the current and adjust it until equilibrium is reached again. Record the added mass of the brass ball(s) and the current.
- Repeat the previous two steps with different added masses in brass balls.

Part VI: Magnetic Resonance
- Turn off the current, turn on the air, and remove the balance beam.
- Place the rotating magnetic field over the air bearing.
- Place the cue ball in the air bearing without the rod.
- Turn on the current to maxmimum.
- Spin the cue ball with high frequency such that the handle is pointing as vertical as it can be.
- Once the ball is spinning, move the rotating magnet such that the handle points through the uprights of the magnet (the handle should never point into the wood). Continue moving the magnet until the spin-flip effect can be observed.
- Record the last step on video to have a recording of the spin-flip effect.

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy.odr as odr
```

```python
def linear(param,x):
    return param[0]*x + param[1]

def correct_radius(radii,d_ball,l_handle,l_weight):
    # converts the data in my list "radii" to the actual distance between the center of the ball to the center of the weight
    for i in range(0,len(radii)):
        radii[i] += 0.5*d_ball + l_handle + 0.5*l_weight

def find_mom_inertia(m,r,m_u,r_u):
    # finds moment of inertia of the ball and the uncertainty in the moment of inertia of the ball
    mom_inertia = 0.4*m*(r**2)
    r2_u = 2*r*r_u
    mom_inertia_u = mom_inertia*(m_u/m + r2_u/(r**2))
    return (mom_inertia,mom_inertia_u)

def curr_to_mag(currents,N,mu_0,R,z,I_u,R_u,z_u):
    # returns the magnetic field and uncertainty in magnetic field associated with each current
    B = []
    B_u = []
    
    for i in range(0,len(currents)):
        mag = (N*mu_0*currents[i]/(2*R))
        B.append(mag)
        mag_u = mag*(R_u/R + I_u/currents[i])
        B_u.append(mag_u)
        
    return (B,B_u)

def find_angles(x,z,x_u,z_u):
    # uses the distance from the center of ball to top of upper coil and
    # horizontal distance rod travels from vertical position to its new position
    # to calculate the angular displacement of the ball/rod [radians]
    angles = []
    angles_u = []
    for i in range(0,len(x)):
        frac = x[i]/z
        angle = math.atan(frac)
        angles.append(angle)
        
        frac_u = (frac)*(x_u[i]/x[i] + z_u/z)
        angle_u = 1/(1+frac**2)*frac_u
        angles_u.append(angle_u)
    return (angles,angles_u)

def recip_of_B(B,B_u):
    # finds the reciprocals of the magnetic fields and their uncertainties
    recip_B = []
    recip_B_u = []
    for i in range(0,len(B)):
        recip_B.append(1/B[i])
        recip_B_u.append(B_u[i]/(B[i]**2))
    return (recip_B,recip_B_u)

def find_periods_squared(t,t_u):
    # given total times of oscillation, finds the periods of oscillations and their uncertainties
    T_squared = []
    T_squared_u = []
    T_u = t_u/20
    for i in range(0,len(t)):
        period = t[i]/20
        T_squared.append(period**2)
        T_squared_u.append(2*period*T_u)
    return (T_squared, T_squared_u)

def find_mag_mom_unc_III(mag_mom,mom_inertia,slope,mom_inertia_u,slope_u):
    # calculates uncertainty in magnetic moment in part III of the experiment
    numer = 4*(math.pi**2)*mom_inertia
    numer_u = 4*(math.pi**2)*mom_inertia_u
    denom_u = slope_u
    unc = mag_mom*(numer_u/numer + denom_u/(5*slope))
    return unc    

def find_freq(periods,T_u):
    # finds precession frequencies and their uncertainties given the periods of precession and their uncertainty
    freqs = []
    freqs_u = []
    for i in range(0,len(periods)):
        freq = 2*math.pi/periods[i]
        freqs.append(freq)
        freq_u = freq*T_u/periods[i]
        freqs_u.append(freq_u)
    return (freqs, freqs_u)

def mass_to_weight(masses,g):
    # converts masses to weights
    weights = []
    for i in range(0,len(masses)):
        weights.append(masses[i]*g)
    return weights

def curr_to_grad(currents,N,mu_0,r,d,I_u,r_u,d_u):
    # converts currents to magnetic gradient strengths and the uncertainties in magnetic gradient strengths
    dB = []
    dB_u = []
    for i in range(0,len(currents)):
        '''numer = 3*mu_0*currents[i]*d*(r**2)
        sum_ = r**2 + ((d**2)/4)
        denom = 2*(sum_)**2.5
        grad = (numer/denom)
        dB.append(grad)
        
        r2_u = 2*r*r_u
        numer_u = numer*(I_u/currents[i] + d_u/d + r2_u/(r**2))
        d2_u = 2*d*d_u
        term2_u = 0.25*d2_u
        sum_u = r2_u + term2_u
        denom_u = 5*(sum_**1.5)*sum_u
        unc = grad*(numer_u/numer + denom_u/denom)
        dB_u.append(unc)'''
        
        dB.append(0.0169*currents[i])
        dB_u.append(0.0169*currents[i])
    
    return (dB,dB_u)
```

```python
# ---------------------------- Defining Useful Variables --------------------------------
m_ball = 157-15.6   # mass of cue ball [kg]
m_ball_u = 0.00015  # uncertainty in mass of cue ball [kg]
m_weight = 0.0014   # mass of weight added in Parts I and II [kg]
m_weight_u = 0.0001 # uncertainty in mass of weight [kg]

d_ball = 0.053      # diameter of cue ball [m]
d_ball_u = 0.001    # uncertainty in diameter of cue ball [m]
l_handle = 0.012    # length of handle on cue ball [m]
l_handle_u = 0.001  # uncertainty in length of handle on cue ball [m]
l_weight = 0.009    # height of weight added in Parts I and II [m]
l_weight_u = 0.001  # uncertainty in height of weight added in Parts I and II [m]
d_cue_top = 0.5*d_ball + 0.045   # distance from center of cue ball to top of upper coil
d_cue_top_u = d_ball_u + 0.0002  # uncertainty in distance from center of cue ball to top of upper coil

N = 390      # number of turns in loop of wire
mu_0 = 1.2566*math.pow(10,-6)     # vacuum permeability [N/A^2]
R = 0.13     # radius of loop of wire [m]
R_u = 0.001  # uncertainty in radius of loop of wire [m]

# moment of inertia of the cue ball and its uncertainty
mom_inertia, mom_inertia_u = find_mom_inertia(m_ball,0.5*d_ball,m_ball_u,0.5*d_ball_u)
```

```python
# ----------------------------------------------- Part I --------------------------------------------------
# Data I collected along with uncertainties in measurements
r = np.array([0,0.005,0.010,0.015,0.020,0.025,0.030]) # distances from tip of handle to center of weight [m]
correct_radius(r,d_ball,l_handle,l_weight) # corrects list "r" to have values for distances from center of cue ball to center of weight [m]
r_u = 0.001    # uncertainty in distance between center of cue ball and center of weight [m]
I = [1.8,1.95,2.0,2.2,2.35,2.50,2.65]  # currents measured [A]
I_u = 0.05   # uncertainties in currents measured [A]
#calculated magnetic field strengths and their uncertainties [T]
(B,B_u) = curr_to_mag(I,N,mu_0,R,d_cue_top,I_u,R_u,d_cue_top_u)

# Plot the data
plt.scatter(r,B)
plt.errorbar(r,B,yerr=B_u,xerr=r_u,fmt="none",color='red',elinewidth=2,linewidth=0,marker='.',markersize=10,label='Error')
plt.xlabel('Radius [m]')
plt.ylabel('Magnetic Field Strength [T]')
plt.title('Figure 1.1: Static Magnetic Torque')
plt.ylim(0.002,0.006)
plt.xlim(0.03,0.08)

# Find a Linear Fit for the data
linear_fit = odr.Model(linear)          
noised_data = odr.RealData(r,B,sx = r_u,sy = B_u)  
regressed_model = odr.ODR(noised_data,linear_fit,beta0=[1,1])  
regressed_output = regressed_model.run() 
regressed_output.pprint()   

X = np.linspace(0,0.10,7)

fitted_params = regressed_output.beta    
param_sd = regressed_output.sd_beta    
chi_squared = str(round(regressed_output.res_var,3))  
fitted_curve = linear(fitted_params,X) 

plt.plot(X,fitted_curve,'orange',label='Linear Fit')

#Show the plot with the linear fit
plt.show()

# Calculate the magnetic moment
mag_mom1 = m_weight / fitted_params[0]
mag_mom_u1 = mag_mom1*(m_weight_u/m_weight + param_sd[0]/fitted_params[0])

# Print the chi-squared of linear fit and the magnetic moment with its uncertainty
print("Chi Squared:", chi_squared)
print("Magnetic Moment: \u03BC =", round(mag_mom1,3), "A/m^2 \u00B1", round(mag_mom_u1,4), "A/m^2")
```

To find magnetic moment in this part of the experiment, I found the strengths of the magnetic field acting on the magnet in the cue ball using the currents in the loops and Biot-Savart Law. Ths slope of the linear fit is a function of magnetic moment, so from there I just solved for magnetic moment.

```python
# ---------------------------------------------- Part III ------------------------------------------------
# Data I collected
I = [1.10,1.30,1.50,1.70,1.90,2.10,2.50,3.00,3.50,4.00,1.20]  # currents [A]
t = [47.34,40.34,35.16,31.91,29.06,27.31,24.09,21.56,19.69,18.32,43.68]  # total times of oscillation over 20 cycles [s]
t_u = 0.5  # uncertainty
# Calculated periods squared with uncertainties and magnetic field strengths with uncertainties
(T_squared, T_squared_u) = find_periods_squared(t,t_u) # [s^2]
(B, B_u) = curr_to_mag(I,N,mu_0,R,d_cue_top,I_u,R_u,d_cue_top_u) # [T]
(recip_B, recip_B_u) = recip_of_B(B,B_u) # Calculates reciprocrals of magnetic field strengths [1/T]
recip_B = np.array(recip_B)

# Plot the data
plt.scatter(recip_B,T_squared)
plt.errorbar(recip_B,T_squared,yerr=T_squared_u,xerr=recip_B_u,fmt="none",color='red',elinewidth=2,linewidth=0,marker='.',markersize=10,label='Error')
plt.xlabel('Reciprocal of Magnetic Field Strength [1/T]')
plt.ylabel('Period Squared [s^2]')
plt.title('Figure 3.1: Harmonic Oscillation')
plt.xlim(0,600)

# Find a linear fit for the data
linear_fit = odr.Model(linear)          
noised_data = odr.RealData(recip_B,T_squared,sx = recip_B_u,sy = T_squared_u)  
regressed_model = odr.ODR(noised_data,linear_fit,beta0=[1,-1])  
regressed_output = regressed_model.run() 
regressed_output.pprint()   

X = np.linspace(50,850,11)

fitted_params = regressed_output.beta    
param_sd = regressed_output.sd_beta    
chi_squared = str(round(regressed_output.res_var,3))  
fitted_curve = linear(fitted_params,X) 

plt.plot(X,fitted_curve,'orange',label='Linear Fit')

# Show the plot with the linear fit
plt.show()

# Calculate magnetic moment with uncertainty
mag_mom3 = (4*(math.pi**2)*mom_inertia)/fitted_params[0]
mag_mom_u3 = find_mag_mom_unc_III(mag_mom3,mom_inertia,fitted_params[0],mom_inertia_u,param_sd[0])

# Print chi-squared value of linear fit and magnetic moment with uncertainty
print("Chi Squared:", chi_squared)
print("Magnetic Moment: \u03BC =", round(mag_mom3,0), "A/m^2 \u00B1", round(mag_mom_u3,1), "A/m^2\n")
```

Precession frequencies were calculated by knowing the periods of precession, and frequency of precession is two times pi divided by period of precession.
